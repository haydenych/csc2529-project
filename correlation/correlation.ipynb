```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from glob import glob

import os
```

```python
def print_noise_map(map):
    assert(len(map.shape) == 2)

    for row in map:
            for num in row:
                print(format(num, ".4f"), end="\t")
            print()
```

```python
def plot_noise_map(map):
    sns.heatmap(map)
```

```python
with open("./output/noise_total_10_320.npy", "rb") as f:
    noise_map_total = np.load(f)
```

```python
print_noise_map(noise_map_total)
```

```python
print_noise_map(np.log(noise_map_total))
```

```python
plot_noise_map(noise_map_total)
```

```python
plot_noise_map(np.log(noise_map_total))
```
